```python
GPT_CONFIG_124M = {
    "vocab_size": 50257,    # Ïñ¥ÌúòÏÇ¨Ï†Ñ ÌÅ¨Í∏∞
    "context_length": 1024, # Î¨∏Îß• Í∏∏Ïù¥
    "emb_dim": 768,         # ÏûÑÎ≤†Îî© Ï∞®Ïõê
    "n_heads": 12,          # Ïñ¥ÌÖêÏÖò Ìó§Îìú Í∞úÏàò
    "n_layers": 12,         # Ï∏µ Í∞úÏàò
    "drop_rate": 0.1,       # ÎìúÎ°≠ÏïÑÏõÉ ÎπÑÏú®
    "qkv_bias": False       # ÏøºÎ¶¨, ÌÇ§, Í∞íÏùÑ ÎßåÎì§ Îïå Ìé∏Ìñ• Ìè¨Ìï® Ïó¨Î∂Ä
}
```

```python
import torch
import torch.nn as nn


class DummyGPTModel(nn.Module):
    def __init__(self, cfg):
        super().__init__()
        self.tok_emb = nn.Embedding(cfg["vocab_size"], cfg["emb_dim"])  # ÌÜ†ÌÅ∞ ÏûÑÎ≤†Îî©
        self.pos_emb = nn.Embedding(cfg["context_length"], cfg["emb_dim"])  # ÏúÑÏπò ÏûÑÎ≤†Îî©
        self.drop_emb = nn.Dropout(cfg["drop_rate"]) # ÎìúÎ°≠ÏïÑÏõÉ

        # Use a placeholder for TransformerBlock
        self.trf_blocks = nn.Sequential( # ÏùºÎ†®Ïùò Ìä∏ÎûòÎÇ¥Ïä§Ìè¨Î®∏ Î∏îÎ°ù
            *[DummyTransformerBlock(cfg) for _ in range(cfg["n_layers"])])

        # Use a placeholder for LayerNorm
        self.final_norm = DummyLayerNorm(cfg["emb_dim"])    # ÏµúÏ¢Ö Ï∏µ Ï†ïÍ∑úÌôî
        self.out_head = nn.Linear(  # ÏÑ†Ìòï Ï∂úÎ†• Ï∏µ
            cfg["emb_dim"], cfg["vocab_size"], bias=False
        )

    def forward(self, in_idx): # Î™®Îç∏ÏùÑ ÌÜµÍ≥ºÌïòÎäî Îç∞Ïù¥ÌÑ∞ ÌùêÎ¶Ñ
        batch_size, seq_len = in_idx.shape
        tok_embeds = self.tok_emb(in_idx)
        pos_embeds = self.pos_emb(torch.arange(seq_len, device=in_idx.device))
        x = tok_embeds + pos_embeds
        x = self.drop_emb(x)
        x = self.trf_blocks(x)
        x = self.final_norm(x)
        logits = self.out_head(x)
        return logits


class DummyTransformerBlock(nn.Module): # Ìä∏ÎûúÏä§ Ìè¨Î®∏ Î∏îÎ°ù
    def __init__(self, cfg):
        super().__init__()
        # ÎçîÎØ∏ ÌÅ¥ÎûòÏä§

    def forward(self, x):
        # Ïù¥ Î∏îÎ°ùÏùÄ ÏïÑÎ¨¥Í≤ÉÎèÑ ÌïòÏßÄ ÏïäÍ≥† ÏûÖÎ†•ÏùÑ Í∑∏ÎÉ• Î∞òÌôòÌï©ÎãàÎã§.
        return x


class DummyLayerNorm(nn.Module): # Ï∏µ Ï†ïÍ∑úÌôî Î∏îÎ°ù
    def __init__(self, normalized_shape, eps=1e-5):
        super().__init__()
        # Ï∏µ Ï†ïÍ∑úÌôî Ïù∏ÌÑ∞ÌéòÏù¥Ïä§Î•º ÌùâÎÇ¥ÎÇ¥Í∏∞ ÏúÑÌïú Îß§Í∞úÎ≥ÄÏàò

    def forward(self, x):
        # Ïù¥ Î∏îÎ°ùÏùÄ ÏïÑÎ¨¥Í≤ÉÎèÑ ÌïòÏßÄ ÏïäÍ≥† ÏûÖÎ†•ÏùÑ Í∑∏ÎÉ• Î∞òÌôòÌï©ÎãàÎã§.
        return x
```

```python
import tiktoken

tokenizer = tiktoken.get_encoding("gpt2")

batch = []

txt1 = "Every effort moves you"
txt2 = "Every day holds a"

batch.append(torch.tensor(tokenizer.encode(txt1)))
batch.append(torch.tensor(tokenizer.encode(txt2)))
batch = torch.stack(batch, dim=0)
print(batch)
```

```python
torch.manual_seed(123)
model = DummyGPTModel(GPT_CONFIG_124M)

logits = model(batch)
print("Ï∂úÎ†• ÌÅ¨Í∏∞:", logits.shape)
print(logits)
```

```python
# Ï∏µ Ï†ïÍ∑úÌôîÎ°ú ÌôúÏÑ±Ìôî Ï†ïÍ∑úÌôîÌïòÍ∏∞
torch.manual_seed(123)

# Îã§ÏÑØ Í∞úÏùò Ï∞®Ïõê(ÌäπÏÑ±)ÏùÑ Í∞ÄÏßÑ Îëê Í∞úÏùò ÌõàÎ†® ÏÉòÌîåÏùÑ ÎßåÎì≠ÎãàÎã§.
batch_example = torch.randn(2, 5)

# Ìï¥Îãπ Ï∏µÏùÄ ÌïòÎÇòÏùò Linear Ï∏µÍ≥º ÎπÑÏÑ†Ìòï ÌôúÏÑ±Ìôî Ìï®ÏàòÏù∏  ReLUÎ°ú Íµ¨ÏÑ±ÎêúÎã§
layer = nn.Sequential(nn.Linear(5, 6), nn.ReLU())
out = layer(batch_example)
print(out)
```

```python
mean = out.mean(dim=-1, keepdim=True)
var = out.var(dim=-1, keepdim=True)

print("ÌèâÍ∑†:\n", mean)
print("Î∂ÑÏÇ∞:\n", var)
# Ï†ïÍ∑úÌôîÎäî Îëê ÏûÖÎ†•(Ìñâ)Ïóê ÎåÄÌï¥ ÎèÖÎ¶ΩÏ†ÅÏúºÎ°ú Ï†ÅÏö©Îê©ÎãàÎã§. dim=-1ÏùÑ ÏÇ¨Ïö©ÌïòÎ©¥ Ìñâ Ï∞®ÏõêÏù¥ ÏïÑÎãàÎùº ÎßàÏßÄÎßâ Ï∞®Ïõê(Ïù¥ Í≤ΩÏö∞ ÌäπÏÑ± Ï∞®Ïõê)ÏùÑ Îî∞Îùº Í≥ÑÏÇ∞Ïù¥ ÏàòÌñâÎê©ÎãàÎã§.
```

```python
# ÌèâÍ∑†ÏùÑ ÎπºÍ≥†, Î∂ÑÏÇ∞Ïùò Ï†úÍ≥±Í∑º(ÌëúÏ§ÄÌé∏Ï∞®)ÏúºÎ°ú ÎÇòÎàÑÎ©¥ ÏûÖÎ†•ÏùÑ Ïó¥(ÌäπÏÑ±) Ï∞®ÏõêÏùÑ Îî∞Îùº ÌèâÍ∑†Ïù¥ 0Ïù¥Í≥† Î∂ÑÏÇ∞Ïù¥ 1Ïù¥ ÎêúÎã§.
out_norm = (out - mean) / torch.sqrt(var)
print("Ï†ïÍ∑úÌôîÎêú Ï∏µ Ï∂úÎ†•:\n", out_norm)

mean = out_norm.mean(dim=-1, keepdim=True)
var = out_norm.var(dim=-1, keepdim=True)
print("ÌèâÍ∑†:\n", mean)
print("Î∂ÑÏÇ∞:\n", var)

# ÌèâÍ∑†Ïùò Ï∂úÎ†•Í∞íÏù∏ 9.9341e-09Îäî 9.9341 * 10**-9 Ïù¥Îã§. Ïù¥Îäî 0Ïóê Îß§Ïö∞ Í∞ÄÍπùÏßÄÎßå 0ÏùÄ ÏïÑÎãàÎã§.
```

```python
# sci_mode = FalseÎ°ú ÏßÄÏ†ïÌïòÏó¨ ÌÖêÏÑúÎ•º Ï∂úÎ†•Ïãú Í≥ºÌïôÏ†Å ÌëúÍ∏∞Î≤ïÏùÑ ÏÇ¨Ïö© Ï†úÌïú, ÏÜåÏàòÏ†¨ ÎÑ§ ÏûêÎ¶¨ÍπåÏßÄÎßå Ï∂úÎ†•
torch.set_printoptions(sci_mode=False)
print("ÌèâÍ∑†:\n", mean)
print("Î∂ÑÏÇ∞:\n", var)
```

```python
class LayerNorm(nn.Module):
    def __init__(self, emb_dim):
        super().__init__()
        self.eps = 1e-5 # Ï†ïÍ∑úÌôîÌï† Îïå 0 ÎÇòÎàóÏÖàÏùÑ Î∞©ÏßÄÌïòÍ∏∞ ÏúÑÌïú Î∂ÑÏÇ∞Ïóê ÎçîÌï¥ÏßÄÎäî ÏûëÏùÄ ÏÉÅÏàò(Ïó°Ïã§Î°†)
        self.scale = nn.Parameter(torch.ones(emb_dim)) # scaleÍ≥º shiftÎäî ÏûÖÎ†•Í≥º Ï∞®ÏõêÏù¥ ÎèôÏùºÌïú ÌõàÎ†® Í∞ÄÎä• ÌååÎùºÎØ∏ÌÑ∞
        self.shift = nn.Parameter(torch.zeros(emb_dim)) # ÌõàÎ†® Í≥ºÏ†ïÏóêÏÑú Îëê ÌååÎùºÎØ∏ÌÑ∞Î•º Ï°∞Ï†ïÌïòÎäî Í≤ÉÏù¥ ÌõàÎ†® ÏûëÏóÖÏóêÏÑú Î™®Îç∏Ïùò ÏÑ±Îä•ÏùÑ Ìñ•ÏÉÅÏãúÌÇ®Îã§Í≥† ÌåêÎã®ÌïòÎäî Í≤ΩÏö∞ LLMÏù¥ ÏûêÎèôÏúºÎ°ú Ï°∞Ï†ï
        # Í∑ºÎç∞ Ïù¥ ÌëúÏ§ÄÌôîÎäî ÎÑàÎ¨¥ Í∞ïÌï¥ÏÑú, Î™®Îç∏Ïù¥ ÏõêÎûò ÌïÑÏöîÎ°ú ÌïòÎäî ÌëúÌòÑ(Ïòà: ÌäπÏ†ï Ï∞®ÏõêÏùÄ ÌÅ¨Í≤å, ÌäπÏ†ï Ï∞®ÏõêÏùÄ ÏûëÍ≤å)ÏùÑ Î¨¥Ï°∞Í±¥ ÏñµÏßÄÎ°ú ÎßûÏ∂∞Î≤ÑÎ¶∞Îã§.
        # Í∑∏ÎûòÏÑú **Œ≥(scale)ÏôÄ Œ≤(shift)**Î•º Î∂ôÏó¨ÏÑú: ùë¶ = ùõæ ‚ãÖ norm_x + ùõΩ
        # y=Œ≥‚ãÖnorm_x+Œ≤
        # Œ≥(scale): Í∞Å Ï∞®Ïõê(feature)Î≥ÑÎ°ú ‚ÄúÏñºÎßàÎÇò ÌÇ§Ïö∏ÏßÄ/Ï§ÑÏùºÏßÄ‚Äù Îã§Ïãú Ï°∞Ï†à
        # Œ≤(shift): Í∞Å Ï∞®Ïõê(feature)Î≥ÑÎ°ú ‚ÄúÏñºÎßàÎÇò Ïù¥ÎèôÌï†ÏßÄ‚Äù Îã§Ïãú Ï°∞Ï†à
        # LayerNormÏù¥ ‚ÄúÏ†ïÎ¶¨‚ÄùÎ•º Ìï¥Ï§Ä Îã§Ïùå, Œ≥/Œ≤Í∞Ä ‚ÄúÌïÑÏöîÌïòÎ©¥ ÏõêÎûò Î∂ÑÌè¨Î°ú Îã§Ïãú ÎêòÎèåÎ¶¥ ÏàòÎèÑ ÏûàÍ≤å‚Äù ÌïòÎäî Ï°∞Ï†à ÎÖ∏Î∏åÎ•º Îã¨ÏïÑÏ£ºÎäî ÏùòÎØ∏
        # LayerNorm / BatchNorm / GroupNorm Í∞ôÏùÄ Ï†ïÍ∑úÌôî Í≥ÑÏó¥ÏóêÎäî Î≥¥ÌÜµ Œ≥/Œ≤Í∞Ä Ï°¥Ïû¨

    def forward(self, x):
        mean = x.mean(dim=-1, keepdim=True)
        var = x.var(dim=-1, keepdim=True, unbiased=False) # correction=0
        # unbiased=False -> Î∂ÑÏÇ∞ Í≥ÑÏÇ∞ Ïãú ÏûÖÎ†•Ïùò Í∞úÏàò nÏúºÎ°ú ÎÇòÎààÎã§. Ïù¥ Î∞©ÏãùÏùÄ ÌëúÎ≥∏ Î∂ÑÏÇ∞ Ï∂îÏ†ïÏùò Ìé∏Ìñ•ÏùÑ Ï°∞Ï†ïÌïòÍ∏∞ ÏúÑÌï¥ n-1Î°ú ÎÇòÎàÑÎäî Î≤†ÏÖÄ Î≥¥Ï†ïÏùÑ Ï†ÅÏö©ÌïòÏßÄ ÏïäÎäîÎã§. Ï∞®Ïõê nÏù¥ ÌÅ¨Îã§Î©¥ nÍ≥º n-1Ïùò Ï∞®Ïù¥Î•º Î¨¥Ïãú Í∞ÄÎä•ÌïòÏßÄÎßå gpt-2 Î™®Îç∏Ïùò Ï†ïÍ∑úÌôîÏôÄ Ìò∏ÌôòÏÑ±ÏùÑ ÏúÑÌï¥ Ìï¥Îãπ Î∞©Ïãù ÏÇ¨Ïö©
        norm_x = (x - mean) / torch.sqrt(var + self.eps)
        return self.scale * norm_x + self.shift
```

```python
ln = LayerNorm(emb_dim=5)
out_ln = ln(batch_example)
```

```python
mean = out_ln.mean(dim=-1, keepdim=True)
var = out_ln.var(dim=-1, unbiased=False, keepdim=True)

print("ÌèâÍ∑†:\n", mean)
print("Î∂ÑÏÇ∞:\n", var)
```

```python
# GELU ÌôúÏÑ±Ìôî Ìï®ÏàòÎ•º ÏÇ¨Ïö©ÌïòÎäî ÌîºÎìú Ìè¨ÏõåÎìú ÎÑ§Ìä∏ÏõåÌÅ¨ Íµ¨ÌòÑÌïòÍ∏∞
class GELU(nn.Module):
    def __init__(self):
        super().__init__()

    def forward(self, x):
        return 0.5 * x * (1 + torch.tanh( # ÏõêÎ≥∏ GPT-2 Î™®Îç∏ÎèÑ Ïª§Î∏å ÌîºÌåÖ(curve fitting)ÏúºÎ°ú Ï∞æÏùÄ Í∑ºÏÇ¨Ïãù ÏÇ¨Ïö©, Í≥ÑÏÇ∞ ÏÜçÎèÑÎ•º ÎÜíÏù¥Í∏∞ ÏúÑÌï¥ Îã§ÏùåÍ≥º Í∞ôÏùÄ Tanh Í∏∞Î∞ò Í∑ºÏÇ¨Ïãù
            torch.sqrt(torch.tensor(2.0 / torch.pi)) *
            (x + 0.044715 * torch.pow(x, 3))
        ))
```

```python
import matplotlib.pyplot as plt

gelu, relu = GELU(), nn.ReLU()

# ÏÉòÌîå Îç∞Ïù¥ÌÑ∞
x = torch.linspace(-3, 3, 100)
y_gelu, y_relu = gelu(x), relu(x)

plt.figure(figsize=(8, 3))
for i, (y, label) in enumerate(zip([y_gelu, y_relu], ["GELU", "ReLU"]), 1):
    plt.subplot(1, 2, i)
    plt.plot(x, y)
    plt.title(f"{label} activation function")
    plt.xlabel("x")
    plt.ylabel(f"{label}(x)")
    plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# FFN
class FeedForward(nn.Module):
    def __init__(self, cfg):
        super().__init__()
        self.layers = nn.Sequential(
            nn.Linear(cfg["emb_dim"], 4 * cfg["emb_dim"]),
            GELU(),
            nn.Linear(4 * cfg["emb_dim"], cfg["emb_dim"]),
        )

    def forward(self, x):
        return self.layers(x)
```

```python
#  ÏàèÏª∑ Ïó∞Í≤∞ Ï∂îÍ∞ÄÌïòÍ∏∞
class ExampleDeepNeuralNetwork(nn.Module):
    def __init__(self, layer_sizes, use_shortcut):
        super().__init__()
        self.use_shortcut = use_shortcut
        self.layers = nn.ModuleList([   # Î†àÏù¥Ïñ¥Ïùò Í∏∞ÌïëÏóê Îî∞Î•∏ Í∑∏Î†àÎîîÏñ∏Ìä∏ ÏÜåÏã§ ÌôïÏù∏Ïö© Îç∞Î™® ÏΩîÎìú
            nn.Sequential(nn.Linear(layer_sizes[0], layer_sizes[1]), GELU()), # ÏÑ†ÌòïÎ≥ÄÌôò Ìïú Î≤à ÌïòÍ≥† ÌôúÏÑ±ÌôîÌï®Ïàò Ï†ÅÏö©
            nn.Sequential(nn.Linear(layer_sizes[1], layer_sizes[2]), GELU()),
            nn.Sequential(nn.Linear(layer_sizes[2], layer_sizes[3]), GELU()),
            nn.Sequential(nn.Linear(layer_sizes[3], layer_sizes[4]), GELU()),
            nn.Sequential(nn.Linear(layer_sizes[4], layer_sizes[5]), GELU())
        ])
        # ModuleListÎäî ÌååÏù¥ÌÜ†ÏπòÍ∞Ä ‚ÄúÏñòÎÑ§ ÏïàÏóê ÌååÎùºÎØ∏ÌÑ∞Í∞Ä ÏûàÎã§‚ÄùÍ≥† Îì±Î°ùÌï¥Ï£ºÎäî Ïª®ÌÖåÏù¥ÎÑàÎùºÏÑú optimizerÍ∞Ä Í∞ÄÏ§ëÏπò ÏóÖÎç∞Ïù¥Ìä∏Î•º Ìï† Ïàò ÏûàÏùå.

    def forward(self, x):
        for layer in self.layers:
            # ÌòÑÏû¨ Ï∏µÏùò Ï∂úÎ†•ÏùÑ Í≥ÑÏÇ∞Ìï©ÎãàÎã§.
            layer_output = layer(x)
            # ÏàèÏª∑ Ïó∞Í≤∞ÏùÑ Ï†ÅÏö©Ìï† Ïàò ÏûàÎäîÏßÄ ÌôïÏù∏Ìï©ÎãàÎã§.
            if self.use_shortcut and x.shape == layer_output.shape: # ÏàèÏª∑ÏùÑ Ïº∞ÏúºÎ©¥(use_shortcut=True) ÏûÖÎ†• xÏôÄ Ï∂úÎ†• layer_outputÏùò shapeÏù¥ Í∞ôÏùÑ ÎïåÎßå x + layer_output
                x = x + layer_output
            else:
                x = layer_output
        return x
        # ResidualÏùÄ ÎçîÌïòÍ∏∞ÎùºÏÑú ÌÖêÏÑú Î™®ÏñëÏù¥ Í∞ôÏïÑÏïº Ìï®. ex) (batch, 128) + (batch, 128) ‚úÖ, (batch, 128) + (batch, 256) ‚ùå (Ï∞®Ïõê Ïïà ÎßûÏïÑÏÑú ÏóêÎü¨)
        # Í≥†Î°ú, Ï∞®ÏõêÏù¥ Ïú†ÏßÄÎêòÎäî Íµ¨Í∞ÑÏóêÏÑúÎßå residual Ï†ÅÏö©ÌïòÍ∏∞ ÏúÑÌïú ÏïàÏ†ÑÏû•Ïπò

def print_gradients(model, x): # "ÏïûÏ™Ω Î†àÏù¥Ïñ¥ Í∞ÄÏ§ëÏπòÍπåÏßÄ Í∑∏ÎùºÎîîÏñ∏Ìä∏Í∞Ä ÏñºÎßàÎÇò ÎèÑÎã¨ÌñàÎäîÏßÄ‚ÄùÎ•º Ï∂úÎ†•
    # Ï†ïÎ∞©Ìñ• Í≥ÑÏÇ∞
    output = model(x)
    target = torch.tensor([[0.]]) # [[ ]]Î°ú Ïã∏ÏÑú Î™®Îç∏ Ï∂úÎ†• Î™®Ïñë(1,1)Í≥º ÎßûÏ∂ò Í≤É

    # ÌÉÄÍπÉÍ≥º Ï∂úÎ†•Ïùò Í∞ÄÍπåÏö¥ Ï†ïÎèÑÎ•º Í∏∞Î∞òÏúºÎ°ú ÏÜêÏã§ÏùÑ Í≥ÑÏÇ∞Ìï©ÎãàÎã§.
    loss = nn.MSELoss()
    loss = loss(output, target)

    # Í∑∏Î†àÏù¥ÎîîÏñ∏Ìä∏Î•º Í≥ÑÏÇ∞ÌïòÍ∏∞ ÏúÑÌïú Ïó≠Ï†ÑÌåå
    loss.backward()

    for name, param in model.named_parameters(): # Linear Î†àÏù¥Ïñ¥Îäî Î≥¥ÌÜµ ÌååÎùºÎØ∏ÌÑ∞Í∞Ä Îëê Í∞ú, ...weight (Í∞ÄÏ§ëÏπò ÌñâÎ†¨), ...bias (Ìé∏Ìñ• Î≤°ÌÑ∞)
        if 'weight' in name:
            # Í∞ÄÏ§ëÏπòÏùò Í∑∏Î†àÏù¥ÎîîÏñ∏Ìä∏Ïùò ÌèâÍ∑† Ï†àÎåìÍ∞íÏùÑ Ï∂úÎ†•Ìï©ÎãàÎã§.
            print(f"{name}Ïùò ÌèâÍ∑† Í∑∏Î†àÏù¥ÎîîÏñ∏Ìä∏Îäî {param.grad.abs().mean().item()}ÏûÖÎãàÎã§.")
```

```python
# ÏàèÏª∑ Ïó∞Í≤∞Ïù¥ ÏóÜÏùÑ Îïå Í∑∏Î†àÎîîÏñ∏Ìä∏ Í∞í
layer_sizes = [3, 3, 3, 3, 3, 1]

sample_input = torch.tensor([[1., 0., -1.]])

torch.manual_seed(123)
model_without_shortcut = ExampleDeepNeuralNetwork(
    layer_sizes, use_shortcut=False
)
print_gradients(model_without_shortcut, sample_input)
```

```python
# **TransformerBlock = (LayerNorm ‚Üí Attention ‚Üí Dropout ‚Üí Residual) (LayerNorm ‚Üí FFN ‚Üí Dropout ‚Üí Residual)**
from previous_chapters import MultiHeadAttention


class TransformerBlock(nn.Module):
    def __init__(self, cfg): # cfgÎäî ÌïòÏù¥ÌçºÌååÎùºÎØ∏ÌÑ∞ Î¨∂Ïùå(dict)
        super().__init__()
        self.att = MultiHeadAttention(
            d_in=cfg["emb_dim"],
            d_out=cfg["emb_dim"],
            context_length=cfg["context_length"],
            num_heads=cfg["n_heads"],
            dropout=cfg["drop_rate"],
            qkv_bias=cfg["qkv_bias"])
        self.ff = FeedForward(cfg) # ÏúÑÏπòÎ≥Ñ(position-wise) MLP
        self.norm1 = LayerNorm(cfg["emb_dim"]) # Attention Ï†ÑÏö© Norm
        self.norm2 = LayerNorm(cfg["emb_dim"]) # FFN Ï†ÑÏö© Norm
        self.drop_shortcut = nn.Dropout(cfg["drop_rate"]) # residual Í≤ΩÎ°ú ÏßÅÏ†ÑÏóê dropout, overfitting Î∞©ÏßÄ, ÌïôÏäµ Ï§ë stochasticity Ï∂îÍ∞Ä

    def forward(self, x):
        # Ïñ¥ÌÖêÏÖò Î∏îÎ°ùÏùÑ ÏúÑÌïú ÏàèÏª∑ Ïó∞Í≤∞
        shortcut = x
        x = self.norm1(x) # Pre-LayerNorm
        x = self.att(x)  # ÌÅ¨Í∏∞: [batch_size, num_tokens, emb_size]
        x = self.drop_shortcut(x)
        x = x + shortcut  # ÏõêÎûò ÏûÖÎ†•ÏùÑ ÎçîÌï©ÎãàÎã§.

        # ÌîºÎìú Ìè¨ÏõåÎìú Î∏îÎ°ùÏùÑ ÏúÑÌïú ÏàèÏª∑ Ïó∞Í≤∞
        shortcut = x
        x = self.norm2(x) # Pre-LayerNorm
        x = self.ff(x)
        x = self.drop_shortcut(x)
        x = x + shortcut  # ÏõêÎûò ÏûÖÎ†•ÏùÑ ÎçîÌï©ÎãàÎã§.

        return x
```

```python
# Ïôú ÏöîÏ¶ò GPT Í≥ÑÏó¥ÏùÄ Pre-LNÎßå Ïì∞ÎÉê?


# Pre-LN
# gradientÍ∞Ä Ï¥àÍ∏∞ Î†àÏù¥Ïñ¥ÍπåÏßÄ ÏïàÏ†ïÏ†ÅÏúºÎ°ú Ï†ÑÎã¨
# ÍπäÏùÄ Î™®Îç∏ÏóêÏÑú ÌïôÏäµ Ïûò Îê®
# ÌïòÏù¥ÌçºÌååÎùºÎØ∏ÌÑ∞ ÌäúÎãù Ïâ¨ÏõÄ

# Post-LN
# ÏñïÏùÄ Î™®Îç∏ÏóêÏÑúÎäî OK
# ÍπäÏñ¥ÏßàÏàòÎ°ù ÌïôÏäµ Î∂àÏïàÏ†ï
# warmup, careful init ÌïÑÏöî
```

```python
# GPT Î™®Îç∏ ÎßåÎì§Í∏∞
# Ìä∏ÎûúÏä§Ìè¨Î®∏ Î∏îÎ°ù 12Î≤à Î∞òÎ≥µ == cfg["n_layers"] = 12

class GPTModel(nn.Module):
    def __init__(self, cfg):
        super().__init__()
        self.tok_emb = nn.Embedding(cfg["vocab_size"], cfg["emb_dim"])
        self.pos_emb = nn.Embedding(cfg["context_length"], cfg["emb_dim"])
        self.drop_emb = nn.Dropout(cfg["drop_rate"])

        self.trf_blocks = nn.Sequential(
            *[TransformerBlock(cfg) for _ in range(cfg["n_layers"])])

        self.final_norm = LayerNorm(cfg["emb_dim"])
        self.out_head = nn.Linear(
            cfg["emb_dim"], cfg["vocab_size"], bias=False
        )

    def forward(self, in_idx):
        batch_size, seq_len = in_idx.shape
        tok_embeds = self.tok_emb(in_idx)
        pos_embeds = self.pos_emb(torch.arange(seq_len, device=in_idx.device))
        x = tok_embeds + pos_embeds  # ÌÅ¨Í∏∞ [batch_size, num_tokens, emb_size]
        x = self.drop_emb(x)
        x = self.trf_blocks(x)
        x = self.final_norm(x)
        logits = self.out_head(x)
        return logits
```

```python
# 4.7 ÌÖçÏä§Ìä∏ ÏÉùÏÑ±ÌïòÍ∏∞
# Ï∂úÎ†• ÌÖêÏÑúÎ•º ÎîîÏΩîÎî©ÌïúÎã§ = Î™®Îç∏Ïù¥ ÎÇ∏ ‚ÄúÎã§Ïùå ÌÜ†ÌÅ∞Ïóê ÎåÄÌïú ÌôïÎ•† Î∂ÑÌè¨‚ÄùÏóêÏÑú ÌÜ†ÌÅ∞ÏùÑ ÌïòÎÇò ÏÑ†ÌÉùÌïòÍ≥†, Í∑∏ ÌÜ†ÌÅ∞ IDÎ•º Ïã§Ï†ú ÌÖçÏä§Ìä∏(Î¨∏Ïûê/Îã®Ïñ¥ Ï°∞Í∞Å)Î°ú Î∞îÍæ∏Îäî Í≥ºÏ†ï
# Ìïú Î≤àÏóê ÌïòÎÇòÏùò ÌÜ†ÌÅ∞ÏùÑ ÎßåÎìúÎäî Î∞òÎ≥µÏ†ÅÏù∏ Í≥ºÏ†ï

def generate_text_simple(model, idx, max_new_tokens, context_size): # ÌòÑÏû¨ ÌÜ†ÌÅ∞ ÏãúÌÄÄÏä§Î•º Î¨∏Îß•ÏúºÎ°ú ÏÇ¨Ïö©Ìï¥ÏÑú Îã§Ïùå ÌÜ†ÌÅ∞ÏùÑ ÌïòÎÇò ÏòàÏ∏° ‚Üí Îí§Ïóê Î∂ôÏù¥Í∏∞Î•º max_new_tokensÎ≤à Î∞òÎ≥µÌïòÎäî greedy decoding Íµ¨ÌòÑ
    # idxÎäî ÌòÑÏû¨ Î¨∏Îß•Ïù¥ Îã¥Í∏¥ (batch, n_tokens) ÌÅ¨Í∏∞Ïùò Ïù∏Îç±Ïä§ Î∞∞Ïó¥ÏûÖÎãàÎã§. ÌòÑÏû¨ Î¨∏Îß• ÌÜ†ÌÅ∞
    # max_new_tokens ‚Üí ÏïûÏúºÎ°ú Î™á Í∞ú ÌÜ†ÌÅ∞ÏùÑ Îçî ÏÉùÏÑ±Ìï†ÏßÄ
    for _ in range(max_new_tokens):

        # ÌòÑÏû¨ Î¨∏Îß•Ïù¥ Î™®Îç∏Ïù¥ ÏßÄÏõêÌïòÎäî Î¨∏Îß• ÌÅ¨Í∏∞Î•º Ï¥àÍ≥ºÌïòÎ©¥ ÏûòÎùºÎÉÖÎãàÎã§.
        # ÏòàÎ•º Îì§Ïñ¥, LLMÏù¥ 5Í∞ú ÌÜ†ÌÅ∞Îßå ÏßÄÏõêÌïòÍ≥† Î¨∏Îß• ÌÅ¨Í∏∞Í∞Ä 10Ïù¥ÎùºÎ©¥,
        # ÎßàÏßÄÎßâ 5Í∞ú ÌÜ†ÌÅ∞Îßå Î¨∏Îß•ÏúºÎ°ú ÏÇ¨Ïö©Ìï©ÎãàÎã§.
        idx_cond = idx[:, -context_size:]
        # idx = [1, 2, 3, 4, 5, 6, 7], context_size = 5, ‚Üí Î™®Îç∏Ïóê ÎÑ£Îäî Í±¥: [3, 4, 5, 6, 7], GPTÏùò ‚ÄúÏä¨ÎùºÏù¥Îî© ÏúàÎèÑÏö∞‚Äù Í∞úÎÖê

        # ÌååÏù¥Ïç¨ with = Ïª®ÌÖçÏä§Ìä∏ Îß§ÎãàÏ†Ä = ‚ÄúÏù¥ Î∏îÎ°ùÏóê Îì§Ïñ¥Ïò§Îäî ÎèôÏïàÎßå Ïñ¥Îñ§ ÏÑ§Ï†ïÏùÑ ÏºúÍ±∞ÎÇò ÎÅàÎã§‚Äù

        # ÏòàÏ∏°ÏùÑ ÎßåÎì≠ÎãàÎã§.
        with torch.no_grad(): # Ï∂îÎ°†(inference) Ïù¥Îùº gradient ÌïÑÏöî ÏóÜÏùå
            logits = model(idx_cond) # Í∞Å ÏúÑÏπòÎßàÎã§ ‚ÄúÎã§Ïùå ÌÜ†ÌÅ∞ ÌôïÎ•† Ï†êÏàò‚Äù
        # torch.no_grad()Îäî Î™®Îç∏Ïùò ÏòàÏ∏° Í≥ÑÏÇ∞ÏùÄ Í∑∏ÎåÄÎ°ú ÏàòÌñâÌïòÎêò, gradient Í≥ÑÏÇ∞ÏùÑ ÏúÑÌïú Ïó∞ÏÇ∞ Í∏∞Î°ù(autograd graph)ÏùÑ ÎπÑÌôúÏÑ±ÌôîÌïòÎäî ÏÑ§Ï†ïÏù¥Îã§.

        # ÎßàÏßÄÎßâ ÌÉÄÏûÑ Ïä§ÌÖùÎßå ÏÇ¨Ïö©ÌïòÎØÄÎ°ú
        # (batch, n_token, vocab_size)Í∞Ä (batch, vocab_size)Í∞Ä Îê©ÎãàÎã§. ÎßàÏßÄÎßâ ÏúÑÏπòÏùò Ï∂úÎ†•Îßå ÏÇ¨Ïö©
        logits = logits[:, -1, :] # Ïù∏Îç±Ïã±

        # ÌôïÎ•†ÏùÑ ÏñªÍ∏∞ ÏúÑÌï¥ ÏÜåÌîÑÌä∏Îß•Ïä§Î•º Ï†ÅÏö©Ìï©ÎãàÎã§.
        probas = torch.softmax(logits, dim=-1)  # (batch, vocab_size)

        # Í∞ÄÏû• ÎÜíÏùÄ ÌôïÎ•† Í∞íÏùÑ Í∞ÄÏßÑ Ìï≠Î™©Ïùò Ïù∏Îç±Ïä§Î•º ÏñªÏäµÎãàÎã§.
        idx_next = torch.argmax(probas, dim=-1, keepdim=True)  # (batch, 1)
        # Í∞ÄÏû• ÌôïÎ•† ÎÜíÏùÄ ÌÜ†ÌÅ∞ ÌïòÎÇò ÏÑ†ÌÉù, greedy decoding, ÎûúÎç§ÏÑ± x, Ìï≠ÏÉÅ Í∞ôÏùÄ Í≤∞Í≥º

        # ÏÑ†ÌÉùÌïú Ïù∏Îç±Ïä§Î•º ÌòÑÏû¨ ÏãúÌÄÄÏä§Ïóê Ï∂îÍ∞ÄÌï©ÎãàÎã§. Î¨∏Îß•Ïù¥ Ï†êÏ†ê Í∏∏Ïñ¥Ïßê
        idx = torch.cat((idx, idx_next), dim=1)  # (batch, n_tokens+1)

    return idx
```

```python
start_context = "Hello, I am"

encoded = tokenizer.encode(start_context)
print("Ïù∏ÏΩîÎî©Îêú ID:", encoded) # (n_tokens,) ÌòÑÏû¨ 1Ï∞®Ïõê

# TransformerÎäî Ìï≠ÏÉÅ (batch, n_tokens), Ï¶â, Î∞∞Ïπò Ï∞®ÏõêÏù¥ ÏûàÏñ¥Ïïº ÌïúÎã§.
# unsqueeze(0) = 0Î≤à Ï∞®ÏõêÏóê ÌÅ¨Í∏∞ 1ÏßúÎ¶¨ Ï∞®Ïõê Ï∂îÍ∞Ä
encoded_tensor = torch.tensor(encoded).unsqueeze(0) # (4,) ‚Üí (1, 4) , (n_tokens)  ‚Üí (batch=1, n_tokens)
print("encoded_tensor.shape:", encoded_tensor.shape)
```

```python
model.eval() # ÎìúÎ°≠ÏïÑÏõÉ ÎÅÑÍ∏∞
# DropoutÏùÄ ÌïôÏäµ Ï§ë: Í≥ºÏ†ÅÌï© Î∞©ÏßÄÏö©, Ï∂îÎ°† Ï§ë: x ÎûúÎç§ÏÑ± ÌïÑÏöî ÏóÜÏùå
# Ï∂îÎ°† Îïå ÏºúÏ†∏ ÏûàÏúºÎ©¥: Ïã§ÌñâÌï† ÎïåÎßàÎã§ Í≤∞Í≥º Îã¨ÎùºÏßê -> ÏòàÏ∏° Î∂àÏïàÏ†ï
# Í∑∏ÎûòÏÑú: ‚ÄúÏù¥Ï†ú ÌïôÏäµ ÎÅù, ÏòàÏ∏°Îßå Ìï† Í±∞Îã§‚ÄùÎùºÍ≥† ÏïåÎ†§Ï£ºÎäî Ïã†Ìò∏Í∞Ä eval()

out = generate_text_simple(
    model=model,
    idx=encoded_tensor,
    max_new_tokens=6, # ÏÉàÎ°ú ÏÉùÏÑ±Ìï† ÌÜ†ÌÅ∞ Í∞úÏàò
    context_size=GPT_CONFIG_124M["context_length"]
)

print("Ï∂úÎ†•:", out) # ÏûÖÎ†• ÌÜ†ÌÅ∞ Í∞úÏàò = 4, ÏÉùÏÑ± ÌÜ†ÌÅ∞ Í∞úÏàò = 6, Ï¥ù Ï∂úÎ†• ÌÜ†ÌÅ∞ Ïàò = 10
print("Ï∂úÎ†• Í∏∏Ïù¥:", len(out[0]))
```
